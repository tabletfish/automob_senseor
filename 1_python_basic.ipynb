## 3) 제어문 (`if`, `for`, `while`)

조건문과 반복문을 이용해 반복적인 점검 작업을 자동화할 수 있습니다.

```python
current_temp = 76

if current_temp > 90:
    print("경고: 온도가 높습니다")
elif current_temp > 75:
    print("온도는 약간 높은 정상 범위입니다")
else:
    print("온도는 정상입니다")

print("\nRPM 점검:")
for rpm in rpm_samples:
    if rpm > 2000:
        print(rpm, "-> 고부하")
    else:
        print(rpm, "-> 정상 부하")
```

## 4) 함수 (Functions)

함수는 코드를 재사용 가능하게 만들고, 테스트하기 쉽게 해줍니다.

```python
def celsius_to_kelvin(temp_c):
    return temp_c + 273.15


def average(values):
    return sum(values) / len(values)

print("36.5C를 Kelvin으로 변환:", celsius_to_kelvin(36.5))
print("평균 RPM:", average(rpm_samples))
```

## 7) 데이터 다루기 심화: 인덱싱부터 전처리까지

이 섹션에서는 실제 데이터 처리를 위해 꼭 필요한 기초를 연습합니다.

### 이번 섹션 학습 목표
- 배열 인덱싱/슬라이싱 정확히 사용하기
- 조건 기반 필터링(불리언 마스킹) 익히기

```python
import numpy as np

# Sample data for practice
temp_c = np.array([72, 75, 79, 88, 91, 85, 77, 69])
rpm = np.array([900, 1100, 1500, 1800, 2200, 2500, 1700, 1200])

sensor_matrix = np.array([
    [72, 900],
    [75, 1100],
    [79, 1500],
    [88, 1800],
    [91, 2200],
    [85, 2500],
])

print("temp_c:", temp_c)
print("rpm:", rpm)
print("sensor_matrix shape:", sensor_matrix.shape)
```

## 8) 배열 인덱싱과 슬라이싱

핵심 규칙:
- `arr[i]`: i번째 원소
- `arr[a:b]`: a 이상 b 미만 구간
- `arr[::k]`: k 간격으로 선택
- `arr[-1]`: 마지막 원소
- 2차원은 `arr[row, col]` 형태 사용

### 체크 포인트
- 시작 인덱스는 포함, 끝 인덱스는 포함하지 않음
- 음수 인덱스로 뒤에서 접근 가능

```python
print("첫 번째 온도:", temp_c[0])
print("마지막 온도:", temp_c[-1])
print("2~4번째 온도:", temp_c[1:4])
print("2칸 간격 온도:", temp_c[::2])

print("\n행렬에서 [2, 1] 값 (3번째 행 RPM):", sensor_matrix[2, 1])
print("행렬의 첫 번째 열 (온도):", sensor_matrix[:, 0])
print("행렬의 두 번째 열 (RPM):", sensor_matrix[:, 1])

# Practice: Change indexes and predict outputs before running
```

## 9) 불리언 마스킹(조건 필터링)

데이터 처리에서 가장 많이 쓰는 패턴입니다.

- 조건식을 만들고 (`temp_c > 80`)
- 그 조건으로 데이터 추출 (`temp_c[조건]`)

```python
high_temp_mask = temp_c > 80
high_rpm_mask = rpm >= 1800

print("고온 조건 마스크:", high_temp_mask)
print("고온 데이터:", temp_c[high_temp_mask])

print("\n고RPM 조건 마스크:", high_rpm_mask)
print("고RPM 데이터:", rpm[high_rpm_mask])

# Combined condition
alert_mask = (temp_c > 85) & (rpm > 1700)
print("\n경고 조건(온도>85 and RPM>1700):", alert_mask)
print("경고 구간 온도:", temp_c[alert_mask])
print("경고 구간 RPM:", rpm[alert_mask])
```

## 11) shape, axis, reshape, broadcasting

### 핵심 개념
- `shape`: 배열 크기(차원별 길이)
- `axis=0`: 행 방향으로 묶어서 계산(열별 결과)
- `axis=1`: 열 방향으로 묶어서 계산(행별 결과)
- `reshape`: 데이터 개수 유지하며 형태 변경
- broadcasting: 서로 다른 shape의 배열을 규칙에 따라 자동 확장해 연산

```python
print("sensor_matrix shape:", sensor_matrix.shape)
print("열별 평균 [온도평균, RPM평균]:", sensor_matrix.mean(axis=0))
print("행별 평균:", sensor_matrix.mean(axis=1))

flat = sensor_matrix.reshape(-1)
print("\n1차원으로 펼치기:", flat)

# Broadcasting example: apply calibration offset
temp_offset = -1.5
rpm_scale = 1.02

calibrated_temp = temp_c + temp_offset
calibrated_rpm = rpm * rpm_scale

print("\n보정된 온도:", calibrated_temp)
print("보정된 RPM:", calibrated_rpm.astype(int))
```

## 12) 결측치(NaN), 이상치, 정렬

실무 데이터는 비어 있거나(결측치), 비정상적으로 큰 값(이상치)을 포함하는 경우가 많습니다.

기초 처리 순서 예시:
1. 결측치 위치 확인
2. 평균/중앙값 등으로 대체 또는 제거
3. 이상치 조건 정의 후 필터링
4. 정렬 및 상위값 추출

```python
raw_temp = np.array([72.0, 75.0, np.nan, 88.0, 150.0, 85.0, np.nan, 69.0])

print("원본 데이터:", raw_temp)
print("NaN 위치:", np.isnan(raw_temp))

# Fill NaN with mean (ignoring NaN)
filled_temp = np.where(np.isnan(raw_temp), np.nanmean(raw_temp), raw_temp)
print("\nNaN 대체 후:", filled_temp)

# Outlier filtering: keep only <= 100
clean_temp = filled_temp[filled_temp <= 100]
print("이상치 제거 후:", clean_temp)

# Sorting and top-k
sorted_temp = np.sort(clean_temp)
print("정렬 결과:", sorted_temp)
print("상위 2개 값:", sorted_temp[-2:])
```
